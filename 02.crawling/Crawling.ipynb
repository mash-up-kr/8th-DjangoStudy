```python
import requests

# 네이버 웹툰 전체 목록 페이지 URL
URL_TOTAL_LIST = 'https://comic.naver.com/webtoon/weekday.nhn'

# 네이버 웹툰의 목록 페이지에 HTTP GET요청 전송, 결과를 response변수에 저장
response = requests.get(URL_TOTAL_LIST)
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
total_a_list = soup.select('a')
len(total_a_list)
```

```python
class_title_a_list = soup.select('a.title')
len(class_title_a_list)
```

```python
# 타이틀 문자열을 저장할 빈 리스트 선언
title_list = []

# soup.select('a.title')의 결과 리스트를 순회
for a in class_title_a_list:
    # 순회하는 각 항목의 get_text() 메서드 결과를 a_text변수에 할당
    a_text = a.get_text()
    a_href = a['href']
    # a_text변수(순회하는 각 a요소의 내용 텍스트)를 title_list리스트에 추가

    # str.format
    # f-string
    result1 = '{} ({})'.format(a_text, a_href)
    result2 = '{text} ({href})'.format(text=a_text, href=a_href)
    result3 = f'{a_text} (https://comic.naver.com{a_href})'
    print(result3)
```
